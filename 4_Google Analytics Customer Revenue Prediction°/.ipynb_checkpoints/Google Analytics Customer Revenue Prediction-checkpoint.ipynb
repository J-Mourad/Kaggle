```python
import pandas as pd
import numpy as np
import seaborn as sns
import json
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_csv("train.csv", dtype={'fullVisitorId': 'str'})
df_test = pd.read_csv("test.csv", dtype={'fullVisitorId': 'str'})
df_train['channelGrouping'] = df_train['channelGrouping'].astype('category')
df_test['channelGrouping'] = df_test['channelGrouping'].astype('category')
df_train.head()
```

```python
def _temp(d, k):
    if k in d.keys():
        # This is for the column 'trafficSource_adwordsClickInfo'
        if type(d[k]) == dict:
            return str(d[k]).replace('False', '"0"').replace('True', '"1"')
        return d[k]
    else:
        return np.nan
def String_dict_to_columns(df, col):
    # Get the index of the largets value in the column to get the keys to build hew columns
    i = df[df[col] == df[col].max()].index
    try:
        if col == 'trafficSource_adwordsClickInfo':
            df[col] = df[col].str.replace('\'', '\"')
        df[col] = df[col].apply(json.loads)
    except: pass
    for k in df[col][i].values[0].keys():
        df[str(col)+'_'+str(k)] = df[col].apply(_temp, args=((k,)))#lambda x: x[k]
    del df[col]
```

```python
# Execute once
for col in ['device', 'geoNetwork', 'totals', 'trafficSource', 'trafficSource_adwordsClickInfo']:
    String_dict_to_columns(df_train, col)
    print("Train" + col)
df_train.head()
# Execute once
for col in ['device', 'geoNetwork', 'totals', 'trafficSource', 'trafficSource_adwordsClickInfo']:
    String_dict_to_columns(df_test, col)
    print("Test" + col)
df_test.head()
```

```python
df_train['date'] = pd.to_datetime(df_train['date'])
df_train['visitStartTime'] = pd.to_datetime(df_train['visitStartTime'])

df_test['date'] = pd.to_datetime(df_test['date'])
df_test['visitStartTime'] = pd.to_datetime(df_test['visitStartTime'])
```
