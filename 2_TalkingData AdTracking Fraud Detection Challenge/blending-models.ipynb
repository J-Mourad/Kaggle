```python
import pandas as pd
import numpy as np
```

```python
print("Reading the data...\n")
df1 = pd.read_csv('../input/talkingdata-lgb-nextclick/my_lgb_sub_0.9772.csv')
df2 = pd.read_csv('../input/ftrl-with-poor-man-s-random-oversampling/sub_proba.csv')
df3 = pd.read_csv('../input/random-forest/my_sub_randm.csv')
df4 = pd.read_csv('../input/ligth-gbm/my_submission.csv')
print("Reading done.....")
```

```python
models = { 'df1' : {
                    'name':'py-lgbm',
                    'score':97.70,
                    'df':df1 },
          'df2' : {
                    'name':'ftrl',
                    'score':96.19,
                    'df':df2 },
           'df3' : {
                   'name':'rdm_frst',
                    'score':97.12,
                    'df':df3 },
          'df4' : {
                    'name':'lgbm_2',
                    'score':96.86,
                    'df':df4 }
         }
```

```python
df1.head()
```

```python
# Making simple blendings of the models

isa_lg = 0
isa_hm = 0
print("Blending...\n")
for df in models.keys() : 
    isa_lg += np.log(models[df]['df'].is_attributed)
    #isa_hm += 1/(models[df]['df'].is_attributed)
isa_lg = np.exp(isa_lg/4)
#isa_hm = 1/isa_hm

print("Isa log\n")
print(isa_lg[:5])
```

```python
sub_log = pd.DataFrame()
sub_log['click_id'] = df1['click_id']
sub_log['is_attributed'] = isa_lg
sub_log.head()
```

```python
print("Writing...")
sub_log.to_csv('sub_log.csv', index=False, float_format='%.9f')
#sub_hm.to_csv('sub_hm.csv', index=False, float_format='%.9f')
```
